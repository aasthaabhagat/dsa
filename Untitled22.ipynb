```python
#largest number from array
def largest(arr):
  if not arr:
    return "array is empty"

  largest = arr[0]
  for num in arr:
    if num > largest:
      largest = num
  return largest
array = [10,60,90,88,77]
largest_element = largest(array)
print(largest_element)

```

```python
def largest(arr):
  return max(arr)
array = [10,50,90,80]
print(f"largest of array is : ",{largest(array)})
```

```python
#second largest of arr
def find_second_largest(arr):
  if len(arr) < 2:
    return -1
  largest = second_largest = -1
  for num in arr:
    if num > largest:
       second_largest = largest
       largest = num
    elif num > second_largest and num != largest:
      second_largest = num
  return second_largest
array = [10,90,80]
print(find_second_largest(array))

```
