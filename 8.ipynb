```python
# check if no is palindrome or n
n = int(input())
if str(n) == str(n)[::-1]:
  print("true")
else:
  print("false")
```

```python
n = int(input())
original = n
reversed = 0
while n > 0:
  digit = n % 10
  reversed = reversed * 10 + digit
  n//= 10
if original == reversed:
  print("true")
else:
  print("false")
```

```python
#Given a range of numbers, find all the palindrome numbers in the range.
min_val = int(input())
max_val = int(input())
for num in range(min_val,max_val + 1):
  if str(num) == str(num)[::-1]:
    print(num,end=" ")
```

```python
#prime no
n = int(input())
def is_prime(n):
  if n < 2:
    return "no"
  for i in range(2,n):
    if n%i == 0:
      return "no"
  return "yes"
print(is_prime(n))
```

```python
#prime no in range
a,b = int(input()),int(input())
for num in range(a,b+1):
  if num > 1:
    is_prime = True
    for i in range(2,num):
      if num%i == 0:
        is_prime = False
    if is_prime:
        print(num,end=" ")
```

```python
#armstrong number
n = int(input())
k = len(str(n))
armstrong_sum = sum(int(digit)**k for digit in str(n))
print(True if armstrong_sum == n else False)
```
