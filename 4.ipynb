```python
def printnos(n):
    if n < 1:
        return
    printnos(n - 1)
    print(n, end=" ")

# Direct execution
n = 5
printnos(n)
```

```python
from typing import List

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        i = 1
        for j in range(1, len(nums)):
            if nums[j] != nums[i - 1]:
                nums[i] = nums[j]
                i += 1
        return i

# Example usage:
sol = Solution()
nums = [1, 1, 2]  # Example input
result = sol.removeDuplicates(nums)
print(result)  # Output: 2
print(nums[:result])  # Output: [1, 2]
```
