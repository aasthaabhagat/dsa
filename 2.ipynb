```python
#sort array using selection sort in increasing order
def selection_sort(arr):
  for i in range(len(arr)):
    min_index = i
    for j in range(i+1,len(arr)):
      if arr[j] < arr[min_index]:
        min_index = j
    arr[i],arr[min_index] = arr[min_index],arr[i]
arr = [2,1,4,3,5]
selection_sort(arr)
print("sorted array : ",arr)
```

```python
 #buble sort
 def bublesort(arr):
  arr.sort()
  return arr
arr = [1,5,6,8,2,0]
bublesort(arr)
print("sorted array " , arr)
```

```python
def insertsort(arr):

    for i in range(1, len(arr)):

        key = arr[i]

        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key


arr = [4, 1, 3, 9, 7]
insertsort(arr)
print("Sorted array:", arr)
```
