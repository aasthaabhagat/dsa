```python
#find second smallest and second largest element in array
def find_second_smallest_largest(arr):
  unique_arr = list(set(arr))
  if len(unique_arr) < 2:
    return -1,-1
  unique_arr.sort()
  return unique_arr[1],unique_arr[-2]
arr = list(map(int,input("enter number seperated by space").split()))
second_smallest, second_largest = find_second_smallest_largest(arr)
print("second smallest : ", second_smallest)
print("second largest :", second_largest)
```

```python
#reverse array
def reverse_array(arr):
  n = len(arr)
  for i in range(n//2):
    arr[i],arr[n-i-1]=arr[n-i-1],arr[i]
  return arr
arr = list(map(int,input("enter nos seperated by space ").split()))
print("reversed array : ", reverse_array(arr))
```
