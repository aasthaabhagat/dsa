```python
# smallest element from array
```

```python
def smallest(arr):
  smallest = arr[0]
  for num in arr:
    if num < smallest:
      smallest = num
  return smallest
arr1 =[2,4,5,6,1]
print("smallest of arr1 is : ", smallest(arr1))
```

```python
def smallest(arr):
  return min(arr)
arr1 = [7,44,22]
print("smallest of arr1 is ",smallest(arr1))
```

```python
def smallest(arr):
  smallest = arr[0]
  for num in arr:
    if num < smallest:
      smallest = num
  return smallest
arr = list(map(int,input("enter nos separated by space : ").split()))
print("smallest element : ",smallest(arr))
```

```python
def largest(arr):
  largest = arr[0]
  for num in arr:
    if num > largest:
      largest = num
  return largest
arr1 = [1,2,111,20,1111]
print("largest element : ", largest(arr1))
```

```python
def largest(arr):
  return max(arr)
arr = [1,2,111,20,111]
print("largest element : ",largest(arr))
```
