```python
#insertion sort
def insertionsort(arr):
  n = len(arr)
  for i in range(1,n):
    key=arr[i]
    j = i-1
    while j>=0 and arr[j]>key:
      arr[j+1] = arr[j]
      j -= 1
    arr[j+1]=key
  return arr
arr1 = [13,46,24,52,20,9]
a=insertionsort(arr1)
print(*a)
```

```python
#quick sort
def quicksort(arr):
  if len(arr) <=1:
    return arr
  pivot = arr[len(arr)//2]
  left=[x for x in arr if x<pivot]
  middle=[x for x in arr if x==pivot]
  right=[x for x in arr if x>pivot]
  return quicksort(left)+middle+quicksort(right)
arr1 = [13,46,24,52,20,9]
a=insertionsort(arr1)
print(*a)
```
